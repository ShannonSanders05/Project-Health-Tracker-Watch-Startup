# **Property Prices vs State Population visualized using a scatterplot**

```python
plt.figure(figsize=(8,6))
plt.scatter(x=price_population_data['avg_price'], y=price_population_data['population'])
plt.title('Property Prices vs Population')
plt.xlabel('Property Prices')
plt.ylabel('Population')
```

# **Average Property prices visualized in a boxplot**

```python
plt.figure(figsize=(6,8))
sns.boxplot(data=price_data, width=0.2)
plt.xlabel('Average Property Price')
```

# **Average State Income vs Property price visualized in a scatterplot**

```python
plt.figure(figsize=(8,6))
plt.scatter(x=price_income_data['average_income'], y=price_income_data['avg_property_price'])
plt.title('Average State Income vs Property Price')
plt.xlabel('Average State Income')
plt.ylabel('Property Price')
```
